```python
folder = "/home/sauerbach/HPC/MolecularDynamics/milestones/04/static"
files = [os.path.join(folder, f) for f in os.listdir(folder) if 'csv' in f]
files = natsort.natsorted(files)

labels = ['0.1', '0.2', '0.3', '0.4',]

for n, f in enumerate(files):
    names = ['kin', 'pot', 'tot', 'temp']
    df = pd.read_csv(f, sep=r"\s+", header=None, names=names)
    plt.plot(df.tot[::2], label=labels[n])
    # plt.plot(df.kin[::2], label=f.split('/')[-1])
    # plt.plot(df.pot[::2], label=f.split('/')[-1])
    print(np.mean(df.tot))
    print(np.std(df.tot))
plt.legend()
plt.xlim(0, 1000)
```
